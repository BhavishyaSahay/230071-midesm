## Step-by-Step Method Description

---

### Step 1: Build the k-Nearest Neighbour Graph
- **Description:** For the input data X = {x₁, ..., xₙ}, construct a symmetric adjacency matrix W ∈ ℝᴺˣᴺ where Wᵢⱼ = 1 if xᵢ is among the p nearest neighbours of xⱼ, or vice versa; otherwise Wᵢⱼ = 0.
- **Reference:** Equation (4), Section "Gaussian Mixture Model with Local Consistency"
- **Purpose:** This graph captures the local geometric (manifold) structure of the data. Two points connected by an edge are considered "nearby" in intrinsic geometry, and the method will penalise divergent cluster assignments between them.

---

### Step 2: Initialise GMM Parameters
- **Description:** Initialise the mixing proportions π = {π₁,...,πₖ}, means μ = {μ₁,...,μₖ}, and covariances Σ = {Σ₁,...,Σₖ} using standard GMM or random initialisation.
- **Reference:** Section "Background", Equation (1)
- **Purpose:** The EM algorithm requires a starting point. Without good initialisation, it may converge to poor local optima.

---

### Step 3: E-Step — Compute Posterior Responsibilities
- **Description:** For each data point xᵢ and component k, compute the posterior probability P(cₖ|xᵢ) using Bayes' rule, treating current parameters as fixed.
- **Reference:** Equation (7)
- **Purpose:** The posterior P(cₖ|xᵢ) represents the "soft" assignment of point xᵢ to cluster k. This is identical to the standard GMM E-step — the manifold regulariser does not affect it.

---

### Step 4: M-Step — Update Mixing Proportions πₖ
- **Description:** Update πₖ = (1/N) Σᵢ P(cₖ|xᵢ). This is unchanged from standard GMM.
- **Reference:** Equation (9)
- **Purpose:** The mixing proportions reflect the relative size of each cluster. Because the regulariser Q₂(Θ) does not depend on πₖ, this update is identical to vanilla GMM.

---

### Step 5: M-Step — Update Means μₖ with KL Regulariser
- **Description:** Update μₖ by maximising the regularised objective. The update adds a correction term to the standard weighted mean: subtract (λ/2Nₖ) × Σᵢⱼ [P(cₖ|xᵢ) − P(cₖ|xⱼ)](xᵢ − xⱼ)Wᵢⱼ.
- **Reference:** Equation (13), derived from Equation (11)
- **Purpose:** The correction term pulls cluster centres so that neighbouring points (connected in the graph) have more similar posterior distributions. This enforces the local consistency assumption.

---

### Step 6: M-Step — Update Covariances Σₖ with KL Regulariser
- **Description:** Update Σₖ with an analogous correction to the standard scatter matrix update. The correction involves the same posterior-difference weights and outer-product scatter matrices Sᵢ,ₖ = (xᵢ − μₖ)(xᵢ − μₖ)ᵀ.
- **Reference:** Equation (14)
- **Purpose:** Ensures that the shape of each Gaussian component also reflects the local geometry, not just its centre.

---

### Step 7: Iterate Until Convergence
- **Description:** Alternate between E-step (Step 3) and M-step (Steps 4–6) until the regularised log-likelihood objective (Equation 6) changes by less than a tolerance ε, or a maximum number of iterations is reached.
- **Reference:** Section "Model Fitting with EM", Equation (8)
- **Purpose:** EM guarantees monotonic increase of the objective under standard GMM; the regularised version retains a well-defined iterative procedure.

---

### Step 8: Assign Final Cluster Labels
- **Description:** Assign each data point xᵢ to the cluster k* = argmax_k P(cₖ|xᵢ).
- **Reference:** Section "Experiment", Evaluation Metric (Equation 15)
- **Purpose:** Converts soft probabilistic assignments into hard cluster labels for evaluation using clustering accuracy (AC) with the Kuhn-Munkres mapping.

---

## Final Summary

This paper solves the problem of **data clustering when the data lies on a low-dimensional manifold** embedded in a high-dimensional Euclidean space; the authors claim their approach outperforms standard GMM by incorporating a KL-Divergence-based local consistency regulariser into the EM objective, which forces neighbouring points (in the k-NN graph sense) to share similar soft cluster assignments, thereby exploiting the geometric structure that GMM ignores.
