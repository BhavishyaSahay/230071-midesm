## Assumption 1

**Assumption:** The data lies on or near a low-dimensional manifold embedded in the ambient (Euclidean) space. Specifically, the conditional distribution P(c|x) varies *smoothly along the geodesics* of this manifold.

**Why the method needs it:** The entire LCGMM regulariser is constructed to enforce smoothness of P(c|x) over a k-NN graph, which is a discrete approximation of the manifold. If the data does not reside on a manifold — for example, if it fills the ambient space uniformly — then the k-NN graph does not approximate any meaningful geometric structure, and the regulariser introduces noise rather than useful structure.

**Violation scenario:** Uniformly distributed data in high dimensions (e.g., random vectors drawn from a hypercube in ℝ²⁰). In this case, nearest neighbours are no more semantically related than distant points, so the graph-based smoothness penalty becomes harmful rather than helpful.

**Paper reference:** Introduction paragraph ("naturally occurring data may reside on or close to an underlying submanifold") and Section "GMM with Locally Consistent Regulariser" — the local consistency assumption is explicitly stated as the foundation for constructing the regulariser.

## Assumption 2

**Assumption:** Two data points that are close in the k-NN graph (i.e., Wᵢⱼ = 1) should have *similar* posterior distributions P(c|xᵢ) ≈ P(c|xⱼ) over the cluster components.

**Why the method needs it:** The regulariser R = Σᵢⱼ 𝒟ᵢⱼ Wᵢⱼ (Equation 5) directly penalises KL divergence between the posteriors of connected points. If this assumption does not hold — e.g., if the true class boundary passes through a densely connected region of the graph — the regulariser will incorrectly penalise correct, distinct assignments for adjacent points on opposite sides of a class boundary.

**Violation scenario:** Data with thin, interleaved clusters (e.g., two interleaved spirals). Points on the boundary of the spirals are geometrically close but belong to different clusters, so forcing their posteriors to agree via the KL regulariser actively harms clustering.

**Paper reference:** Section "GMM with Locally Consistent Regulariser" — "within some neighbouring samples, their P(c|xᵢ) are similar to a certain degree" (local consistency assumption, also cited as Zhou et al. 2003). Equation (5) operationalises this assumption directly.

## Assumption 3

**Assumption:** The number of clusters K and the number of nearest neighbours p are known or can be set empirically before fitting. The algorithm has no mechanism for automatically selecting these hyperparameters.

**Why the method needs it:** The EM algorithm for GMM requires K to be fixed in advance (it determines the number of Gaussian components), and the graph structure depends entirely on p. The regularisation strength λ is also a free parameter. The LCGMM objective (Equation 6) has no built-in model selection criterion — it is simply maximised for a given (K, p, λ).

**Violation scenario:** Real-world datasets where the number of clusters is unknown (e.g., customer segmentation on a new market). Choosing K incorrectly leads to either over-splitting or under-splitting clusters, and neither the regulariser nor the EM update can correct for this. The paper itself empirically fixes p = 20 and λ = 0.1 for all experiments (Section "Compared Algorithms").

**Paper reference:** Section "Model Selection" and "Compared Algorithms" — both p and λ are fixed across experiments. The paper acknowledges these as two essential parameters but provides only a sensitivity plot (Figure 2) rather than a principled selection method.
