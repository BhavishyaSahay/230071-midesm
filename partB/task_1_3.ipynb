## Main Baseline

The paper's primary baseline is the **standard Gaussian Mixture Model (GMM)** fitted via the Expectation-Maximisation (EM) algorithm (Bishop 2006). The paper also compares against k-means and Normalised Cut (NCut) spectral clustering, but the core claimed improvement is specifically over GMM, since LCGMM is a direct extension of it.

---

## Limitation of the Baseline

Standard GMM assumes that the probability distribution generating the data is supported on **flat Euclidean space**. It treats all data points as being drawn from a mixture of Gaussians in ℝᴰ, and its EM updates compute cluster means and covariances purely through weighted Euclidean averages. This means GMM is **blind to any manifold structure** in the data: if the data lies on a curved submanifold (e.g., a Swiss roll, a face image manifold), GMM's cluster boundaries are shaped only by the Gaussian density in ambient space and do not respect the intrinsic geometry. As a result, GMM produces poor clusters when the data has strong non-Euclidean local structure.

---

## How LCGMM Overcomes This

LCGMM adds a KL-Divergence-based **local consistency regulariser** (Equation 6) to the GMM log-likelihood. This regulariser penalises pairs of graph-adjacent points (determined by a k-NN graph) that receive dissimilar soft cluster assignments. The modified M-step updates for μₖ and Σₖ (Equations 13 and 14) therefore incorporate a correction that encourages geometrically close points to belong to the same cluster, effectively incorporating the manifold structure into the EM procedure without changing the E-step or the form of the Gaussian components.

---

## Condition Under Which LCGMM Would NOT Outperform GMM

LCGMM would fail to outperform standard GMM — and would likely perform worse — when the **true cluster boundaries bisect densely connected regions of the k-NN graph**. If the data has well-separated, roughly Gaussian clusters in ambient space (i.e., the setting that GMM is designed for), the kNN graph will connect points across cluster boundaries. The regulariser will then actively penalise the correct cluster assignments for those boundary points, distorting the parameter updates and reducing accuracy. This is consistent with the paper's own results: on the **Waveform** dataset, LCGMM (75.3%) is outperformed by standard GMM (76.3%), which the authors attribute implicitly to the data being well-described by Euclidean structure. Additionally, for very large λ (over-regularisation), the regularisation term dominates and clusters collapse toward a single assignment for all locally-connected points, which destroys meaningful clustering entirely. The paper's own sensitivity analysis in Figure 2 shows accuracy dropping significantly once λ exceeds ~0.1.
